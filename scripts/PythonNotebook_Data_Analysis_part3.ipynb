# Global Health Trends Analysis (1990â€“2023)

This notebook analyzes global health trends across seven major regions using data from `region_data.csv`. It focuses on three key health indicators:
- **Life Expectancy**
- **Under-5 Mortality Rate**
- **Adolescent Fertility Rate**

The analysis includes:
1. Visualizing trends over time for each indicator.
2. Comparing changes between 1990 and 2023.
3. Calculating percentage changes for each region.
4. Exploring correlations between life expectancy improvements and reductions in under-5 mortality and adolescent fertility rates.

This notebook provides insights into regional disparities, progress over time, and the relationships between key health indicators.

1. Importing Libraries and Data

Insight: This chunk imports the necessary libraries (matplotlib, seaborn, pandas) and loads the dataset region_data.csv into a DataFrame for analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the clean_data.csv file into a DataFrame
region_df = pd.read_csv('../data/region_data.csv')
```

2. Visualizing Life Expectancy Trends

Insight: This chunk visualizes life expectancy trends from 1990 to 2023 for seven regions, highlighting disparities and progress over time.

```python
# Plot Life Expectancy over Time for 7 regions
plt.figure(figsize=(12,6))
sns.lineplot(
    data=region_df,
    x='date',
    y='life_expectancy',
    hue='country',
    marker='o'
)

plt.title('Life Expectancy Trends by Region (1990â€“2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Life Expectancy (Years)', fontsize=14)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

3. Visualizing Under-5 Mortality Trends

Insight: This chunk shows the dramatic decline in under-5 mortality rates across all regions, with Sub-Saharan Africa and South Asia showing significant progress.

```python
# Plot Under-5 Mortality over Time for 7 regions
plt.figure(figsize=(12,6))
sns.lineplot(
    data=region_df,
    x='date',
    y='under5_mortality_rate',
    hue='country',
    marker='o'
)

plt.title('Under-5 Mortality Rate Trends by Region (1990â€“2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Under-5 Mortality Rate (per 1,000 live births)', fontsize=14)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

4. Visualizing Adolescent Fertility Trends

Insight: This chunk highlights the decline in adolescent fertility rates, with South Asia achieving the steepest reductions.

```python
# Plot Adolescent Fertility over Time for 7 regions
plt.figure(figsize=(12,6))
sns.lineplot(
    data=region_df,
    x='date',
    y='adolescent_fertility_rate',
    hue='country',
    marker='o'
)

plt.title('Adolescent Fertility Rate Trends by Region (1990â€“2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Adolescent Fertility Rate (per 1,000 women ages 15â€“19)', fontsize=14)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

5. Comparing 1990 and 2023

Insight: This chunk filters the data for 1990 and 2023, merges the two datasets, and displays the combined data for comparison.

```python
#| echo: false
#| output: asis

# Filter only for 1990 and 2023
df_1990 = region_df[region_df['date'] == 1990]
df_2023 = region_df[region_df['date'] == 2023]

# Merge 1990 and 2023 data by region
region_change = pd.merge(
    df_1990,
    df_2023,
    on='country',
    suffixes=('_1990', '_2023')
)

# Display the table as plain text
print(region_change.head())

# Display nicely for PDF
from IPython.display import display, Latex

# Generate LaTeX table
latex_table = r"""
\begin{table}[!htbp]
\centering
\small
\resizebox{\textwidth}{!}{
""" + region_change.head().to_latex(index=False, escape=True) + r"""
}
\end{table}
"""

# Display the LaTeX table
display(Latex(latex_table))
```

6. Calculating Percentage Changes

Insight: This chunk calculates the percentage changes in life expectancy, under-5 mortality, and adolescent fertility rates for each region between 1990 and 2023.

```python
#| echo: false
#| output: asis

# Calculate % change for each indicator
region_change['life_expectancy_pct_change'] = (
    (region_change['life_expectancy_2023'] - region_change['life_expectancy_1990'])
    / region_change['life_expectancy_1990']
) * 100

region_change['under5_mortality_pct_change'] = (
    (region_change['under5_mortality_rate_2023'] - region_change['under5_mortality_rate_1990'])
    / region_change['under5_mortality_rate_1990']
) * 100

region_change['adolescent_fertility_pct_change'] = (
    (region_change['adolescent_fertility_rate_2023'] - region_change['adolescent_fertility_rate_1990'])
    / region_change['adolescent_fertility_rate_1990']
) * 100

# View final table
region_change[['country', 'life_expectancy_pct_change', 'under5_mortality_pct_change', 'adolescent_fertility_pct_change']]

# Select final table
final_table = region_change[['country', 'life_expectancy_pct_change', 'under5_mortality_pct_change', 'adolescent_fertility_pct_change']]

# Display the table as plain text
print(final_table.head())

# Display nicely for PDF
from IPython.display import display, Latex

display(Latex(r"""
\begin{table}[!htbp]
\centering
\small
\resizebox{\textwidth}{!}{
""" + final_table.to_latex(index=False, escape=True) + r"""
}
\end{table}
"""))
```

7. Visualizing Percentage Changes

Insight: This chunk creates bar plots to visualize the percentage changes for each indicator, highlighting regional disparities and progress.

```python
# Create a combined figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(20,6), sharey=True)

# Plot 1: Life Expectancy % Change
sns.barplot(
    data=region_change.sort_values('life_expectancy_pct_change', ascending=False),
    x='life_expectancy_pct_change',
    y='country',
    ax=axes[0],
    palette='viridis'
)
axes[0].set_title('Life Expectancy % Change', fontsize=14)
axes[0].set_xlabel('% Change')
axes[0].set_ylabel('Region')
axes[0].grid(True)

# Plot 2: Under-5 Mortality % Change
sns.barplot(
    data=region_change.sort_values('under5_mortality_pct_change', ascending=True),
    x='under5_mortality_pct_change',
    y='country',
    ax=axes[1],
    palette='coolwarm'
)
axes[1].set_title('Under-5 Mortality % Change', fontsize=14)
axes[1].set_xlabel('% Change')
axes[1].set_ylabel('')
axes[1].grid(True)

# Plot 3: Adolescent Fertility % Change
sns.barplot(
    data=region_change.sort_values('adolescent_fertility_pct_change', ascending=True),
    x='adolescent_fertility_pct_change',
    y='country',
    ax=axes[2],
    palette='magma'
)
axes[2].set_title('Adolescent Fertility % Change', fontsize=14)
axes[2].set_xlabel('% Change')
axes[2].set_ylabel('')
axes[2].grid(True)

# Main title
fig.suptitle('Population Health Improvements by Region (1990â€“2023)', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

8. Correlation Analysis

Insight: This chunk calculates and prints the correlations between life expectancy improvements and reductions in under-5 mortality and adolescent fertility rates.

```python
# Correlation between life expectancy improvements and other indicators
corr_under5 = region_change[['life_expectancy_pct_change', 'under5_mortality_pct_change']].corr().iloc[0,1]
corr_fertility = region_change[['life_expectancy_pct_change', 'adolescent_fertility_pct_change']].corr().iloc[0,1]

print(f"Correlation between Life Expectancy Change and Under-5 Mortality Change: {corr_under5:.3f}")
print(f"Correlation between Life Expectancy Change and Adolescent Fertility Change: {corr_fertility:.3f}")
```

9. Calculating Correlations

Insight:

Pearson correlation coefficients are calculated to quantify the strength and direction of the relationships:
Between life expectancy improvements and under-5 mortality reductions (corr_under5).
Between life expectancy improvements and adolescent fertility reductions (corr_fertility).

```python
import scipy.stats as stats

# Calculate correlations
corr_under5, _ = stats.pearsonr(region_change['under5_mortality_pct_change'], region_change['life_expectancy_pct_change'])
corr_fertility, _ = stats.pearsonr(region_change['adolescent_fertility_pct_change'], region_change['life_expectancy_pct_change'])
```

10. Scatterplot: Life Expectancy vs Under-5 Mortality Improvements

Insight:

A scatterplot with a regression line visualizes the relationship between life expectancy improvements and under-5 mortality reductions.
Each point represents a region, and the plt.text() function annotates the points with region names.
The correlation coefficient (r = {corr_under5:.2f}) is displayed on the plot to summarize the strength of the relationship.

```python
# Scatterplot: Life Expectancy vs Under-5 Mortality Improvements
plt.figure(figsize=(8,6))
sns.regplot(
    data=region_change,
    x='under5_mortality_pct_change',
    y='life_expectancy_pct_change',
    scatter=True,
    color='teal',
    marker='o'
)
for i in range(region_change.shape[0]):
    plt.text(
        region_change['under5_mortality_pct_change'].iloc[i]+1,
        region_change['life_expectancy_pct_change'].iloc[i]+1,
        region_change['country'].iloc[i],
        fontsize=9
    )
plt.title('Life Expectancy vs Under-5 Mortality Improvements', fontsize=16)
plt.xlabel('Under-5 Mortality % Change')
plt.ylabel('Life Expectancy % Change')
plt.grid(True)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
# Add correlation text
plt.text(
    0.05, 0.95, f'r = {corr_under5:.2f}',
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top'
)
plt.show()
```

11. Scatterplot: Life Expectancy vs Adolescent Fertility Improvements

Insight:

A scatterplot with a regression line visualizes the relationship between life expectancy improvements and adolescent fertility reductions.
Similar to the previous plot, points are annotated with region names, and the correlation coefficient (r = {corr_fertility:.2f}) is displayed.

```python
# Scatterplot: Life Expectancy vs Adolescent Fertility Improvements
plt.figure(figsize=(8,6))
sns.regplot(
    data=region_change,
    x='adolescent_fertility_pct_change',
    y='life_expectancy_pct_change',
    scatter=True,
    color='purple',
    marker='o'
)
for i in range(region_change.shape[0]):
    plt.text(
        region_change['adolescent_fertility_pct_change'].iloc[i]+1,
        region_change['life_expectancy_pct_change'].iloc[i]+1,
        region_change['country'].iloc[i],
        fontsize=9
    )
plt.title('Life Expectancy vs Adolescent Fertility Improvements', fontsize=16)
plt.xlabel('Adolescent Fertility % Change')
plt.ylabel('Life Expectancy % Change')
plt.grid(True)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
# Add correlation text
plt.text(
    0.05, 0.95, f'r = {corr_fertility:.2f}',
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top'
)
plt.show()
```
